Write a SQL query to retrieve all columns for sales made on '2022-11-05:

Write a SQL query to retrieve all transactions where the category is 'Clothing' and the quantity sold is more than 4 in the month of Nov-2022:

```python
query="""SELECT 
  *
FROM retail_sales
WHERE 
    category = 'Clothing'
    
	AND 
    DATE_FORMAT(sale_date, "%Y-%m")='2022-11'
  
    AND
    quantity >=4;
         """
pd.read_sql_query(query,con=connection)
```

Write a SQL query to calculate the total sales (total_sale) for each category.:

```python
query="""SELECT category,SUM(total_sale) AS Total_SALES
         FROM retail_sales
         GROUP BY category
"""
pd.read_sql_query(query,con=connection)
```

Write a SQL query to find the average age of customers who purchased items from the 'Beauty' category.:

```python
query="""SELECT category,ROUND(AVG(age),2) AS AVG_AGE
         FROM retail_sales
         WHERE category='Beauty'
         GROUP BY 1"""

pd.read_sql_query(query,con=connection)
```

Write a SQL query to find all transactions where the total_sale is greater than 1000.:

```python
query="""SELECT  *
         FROM retail_sales
         WHERE total_sale>1000
         """

pd.read_sql_query(query,con=connection)
```

Write a SQL query to find the total number of transactions (transaction_id) made by each gender in each category.:

```python
query="""SELECT  category,gender,COUNT(transactions_id) AS TOTAL_TRANS
         FROM retail_sales
         GROUP BY 1,2
         """

pd.read_sql_query(query,con=connection)
```

Write a SQL query to calculate the average sale for each month. Find out best selling month in each year:

```python
query="""WITH SALES_YEAR_MONTH_WISE AS (
                SELECT 	
                DATE_FORMAT(sale_date, "%Y") AS YEAR, 
                DATE_FORMAT(sale_date, "%M") AS MONTH,
                 ROUND(AVG(total_sale),2) AS TOTAL_SALES_AVG,
                 ROW_NUMBER() OVER(PARTITION BY DATE_FORMAT(sale_date, "%Y") ORDER BY AVG(total_sale) DESC ) AS RN
         FROM retail_sales
         GROUP BY YEAR,MONTH)
	SELECT * FROM SALES_YEAR_MONTH_WISE
    WHERE RN=1
         """

pd.read_sql_query(query,con=connection)
```

**Write a SQL query to find the top 5 customers based on the highest total sales **:

```python
query="""SELECT customer_id,SUM(total_sale) AS TOTAL_SALES
         FROM retail_sales
         GROUP BY customer_id
         ORDER BY  TOTAL_SALES DESC
         LIMIT 5
         """

pd.read_sql_query(query,con=connection)
```

Write a SQL query to find the number of unique customers who purchased items from each category.:

```python
query="""SELECT category,COUNT(DISTINCT CUSTOMER_ID) AS UNIQUE_CUSTOMER
         FROM retail_sales
         GROUP BY category
         """

pd.read_sql_query(query,con=connection)
```

Write a SQL query to create each shift and number of orders (Example Morning <12, Afternoon Between 12 & 17, Evening >17):

```python
query="""    WITH swift_wise AS(
    SELECT *,
    CASE WHEN TIME_FORMAT(sale_time,"%H")<12 THEN 'Morning'
    WHEN TIME_FORMAT(sale_time,"%H")>17 THEN 'Evening'
    ELSE "Afternoon"
    END AS "shift"
    FROM retail_sales)
    SELECT shift,COUNT(transactions_id) AS CUSTOMER_COUNT FROM  swift_wise
    GROUP BY shift
         """

pd.read_sql_query(query,con=connection)
```
